```python
# الخطوة 2: تحميل الموديل العربي
import torch
from transformers import Wav2Vec2Processor, Wav2Vec2Model

device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"🖥️ الجهاز: {device}")

processor = Wav2Vec2Processor.from_pretrained("elgeish/wav2vec2-large-xlsr-53-arabic")
model     = Wav2Vec2Model.from_pretrained("elgeish/wav2vec2-large-xlsr-53-arabic")
model     = model.to(device)
model.eval()
print("✅ الموديل جاهز!")
```

```python
# الخطوة 5: استخراج Embeddings بالموديل العربي
import torch
import librosa
import numpy as np
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity

df = pd.read_csv("features_arabic.csv")
print(f"📊 عدد التسجيلات: {len(df)}")

def preprocess_audio(file_path, target_sr=16000):
    audio, sr = librosa.load(file_path, sr=target_sr)
    audio, _  = librosa.effects.trim(audio, top_db=20)
    return audio

# استخراج الـ Embeddings
print("⏳ جاري استخراج الـ Embeddings...")
embeddings = {}

for i, row in df.iterrows():
    file_path = f"{row['record_id']}.wav"
    try:
        audio  = preprocess_audio(file_path)
        inputs = processor(audio, sampling_rate=16000, return_tensors="pt", padding=True)
        inputs = {k: v.to(device) for k, v in inputs.items()}
        with torch.no_grad():
            outputs = model(**inputs)
        emb = outputs.last_hidden_state.mean(dim=1).squeeze().cpu().numpy()
        embeddings[row['record_id']] = emb
        if (i+1) % 200 == 0:
            print(f"✅ {i+1}/2000")
    except Exception as e:
        print(f"⚠️ {row['record_id']}: {e}")

print(f"\n✅ تم استخراج {len(embeddings)} Embedding!")
```

```python
# الخطوة 6: بناء المراجع وحساب Cosine Similarity
print("⏳ جاري بناء المراجع...")
references = {}
for word in df['word'].unique():
    correct_ids = df[(df['word'] == word) & (df['label'] == 0)]['record_id'].tolist()
    embs = [embeddings[rid] for rid in correct_ids if rid in embeddings]
    if embs:
        references[word] = np.mean(embs, axis=0)

print(f"✅ تم بناء {len(references)} مرجع!")

# حساب Cosine Similarity
print("⏳ جاري حساب Cosine Similarity...")
cos_sims = []
for i, row in df.iterrows():
    word = row['word']
    rid  = row['record_id']
    if word in references and rid in embeddings:
        sim = cosine_similarity(
            embeddings[rid].reshape(1, -1),
            references[word].reshape(1, -1)
        )[0][0]
    else:
        sim = 0.0
    cos_sims.append(round(float(sim), 4))

df['arabic_cosine_v2'] = cos_sims
df.to_csv("features_final.csv", index=False)
print(f"🎉 خلصنا!")
print(f"📊 متوسط Arabic Cosine: {np.mean(cos_sims):.4f}")
```

```python
# الخطوة 7: تدريب XGBoost
from xgboost import XGBClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, accuracy_score

train = df[df["split"] == "train"]
val   = df[df["split"] == "val"]
test  = df[df["split"] == "test"]

features = ["pitch", "intensity", "duration", "arabic_cosine_v2"]

X_train, y_train = train[features].values, train["label"].values
X_val,   y_val   = val[features].values,   val["label"].values
X_test,  y_test  = test[features].values,  test["label"].values

scaler  = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_val   = scaler.transform(X_val)
X_test  = scaler.transform(X_test)

print("⏳ جاري التدريب...")
clf = XGBClassifier(
    n_estimators=500,
    max_depth=4,
    learning_rate=0.03,
    subsample=0.8,
    colsample_bytree=0.8,
    scale_pos_weight=len(y_train[y_train==0])/len(y_train[y_train==1]),
    random_state=42
)
clf.fit(X_train, y_train)
print("✅ التدريب خلص!")

print(f"\n🎯 Val:  {accuracy_score(y_val, clf.predict(X_val)):.2%}")
print(classification_report(y_val, clf.predict(X_val), target_names=["Correct","Mispronounced"]))

print(f"\n🎯 Test: {accuracy_score(y_test, clf.predict(X_test)):.2%}")
print(classification_report(y_test, clf.predict(X_test), target_names=["Correct","Mispronounced"]))
```

```python
# الخطوة 8: حفظ وتنزيل الملفات
import joblib
from google.colab import files

joblib.dump(clf,    "xgboost_final.pkl")
joblib.dump(scaler, "scaler_final.pkl")

files.download("xgboost_final.pkl")
files.download("scaler_final.pkl")
files.download("features_final.csv")

print("✅ تم تنزيل كل الملفات!")
```
